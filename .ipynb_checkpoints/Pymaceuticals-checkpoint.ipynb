```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
clinicaltrial_data = pd.read_csv('clinicaltrial_data.csv')
mouse_drug_data = pd.read_csv('mouse_drug_data.csv')
mouse_drug_data.tail()
clinicaltrial_data.tail()
```

```python
#merge on mouse ID
combined_data = clinicaltrial_data.merge(mouse_drug_data,how='left',on='Mouse ID')
combined_data.head()
```

```python
#Group by treatment/drug
treatment_df = combined_data.drop(['Mouse ID', 'Metastatic Sites'],axis=1)
treatment_df = treatment_df.loc[(treatment_df['Drug']=='Capomulin')| (treatment_df['Drug']=='Infubinol')| (treatment_df['Drug']=='Ketapril')| (treatment_df['Drug']=='Placebo')]

treatment_df = treatment_df.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].mean()
treatment_df = treatment_df.to_frame()
#treatment_df = treatment_df.pivot(index='Timepoint',columns='Drug',values='Tumor Volume (mm3)')
treatment_df = treatment_df.unstack(level=0)
treatment_df = treatment_df.reset_index()
treatment_df
```

```python
cap, = plt.plot(treatment_df['Timepoint'],treatment_df['Tumor Volume (mm3)']['Capomulin'],linewidth=1,label='Capomulin',marker='s')
inf, = plt.plot(treatment_df['Timepoint'],treatment_df['Tumor Volume (mm3)']['Infubinol'],linewidth=1,label='Infubinol',marker='s')
ket, = plt.plot(treatment_df['Timepoint'],treatment_df['Tumor Volume (mm3)']['Ketapril'],linewidth=1,label='Ketapril',marker='s')
pla, = plt.plot(treatment_df['Timepoint'],treatment_df['Tumor Volume (mm3)']['Placebo'],linewidth=1,label='Placebo',marker='s')

plt.title('Tumor Response to treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.grid(True)
plt.legend(handles=[cap,inf,ket,pla],loc='best')
plt.show()

```

```python
metastatic = combined_data.drop(['Mouse ID', 'Tumor Volume (mm3)'],axis=1)
metastatic = metastatic.loc[(metastatic['Drug']=='Capomulin')| (metastatic['Drug']=='Infubinol')| (metastatic['Drug']=='Ketapril')| (metastatic['Drug']=='Placebo')]

metastatic = metastatic.groupby(['Drug','Timepoint'])['Metastatic Sites'].mean()
metastatic = metastatic.to_frame()
metastatic = metastatic.unstack(level=0)
metastatic = metastatic.reset_index()
metastatic
```

```python
cap, = plt.plot(metastatic['Timepoint'],metastatic['Metastatic Sites']['Capomulin'],linewidth=1,label='Capomulin',marker='s')
inf, = plt.plot(metastatic['Timepoint'],metastatic['Metastatic Sites']['Infubinol'],linewidth=1,label='Infubinol',marker='s')
ket, = plt.plot(metastatic['Timepoint'],metastatic['Metastatic Sites']['Ketapril'],linewidth=1,label='Ketapril',marker='s')
pla, = plt.plot(metastatic['Timepoint'],metastatic['Metastatic Sites']['Placebo'],linewidth=1,label='Placebo',marker='s')

plt.title('Metastatic response during treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Metastatic Sites')
plt.grid(True)
plt.legend(handles=[cap,inf,ket,pla],loc='best')
plt.show()
```

```python
mouse = combined_data.drop(['Metastatic Sites', 'Tumor Volume (mm3)'],axis=1)
mouse = mouse.loc[(mouse['Drug']=='Capomulin')| (mouse['Drug']=='Infubinol')| (mouse['Drug']=='Ketapril')| (mouse['Drug']=='Placebo')]

mouse = mouse.groupby(['Drug','Timepoint'])['Mouse ID'].count()
mouse = mouse.to_frame()
mouse = mouse.unstack(level=0)
mouse = mouse.reset_index()
mouse
```

```python
cap, = plt.plot(mouse['Timepoint'],mouse['Mouse ID']['Capomulin'],linewidth=1,label='Capomulin',marker='s')
inf, = plt.plot(mouse['Timepoint'],mouse['Mouse ID']['Infubinol'],linewidth=1,label='Infubinol',marker='s')
ket, = plt.plot(mouse['Timepoint'],mouse['Mouse ID']['Ketapril'],linewidth=1,label='Ketapril',marker='s')
pla, = plt.plot(mouse['Timepoint'],mouse['Mouse ID']['Placebo'],linewidth=1,label='Placebo',marker='s')

plt.title('Survival during treatment')
plt.xlabel('Time (Days)')
plt.ylabel('Survival rate')
plt.grid(True)
plt.legend(handles=[cap,inf,ket,pla],loc='best')
plt.show()
```
