```python
#Import dependencies
import pandas as pd
import csv
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Read into data file and see information provided
city_data = pd.read_csv('city_data.csv')
ride_data = pd.read_csv('ride_data.csv')

print(ride_data.head())
print(city_data.head())
```

```python
#Combine dataframes into one for easier analysis
combined_data = ride_data.merge(city_data, on='city',how='left')
combined_data.tail()
```

```python
#Create dataframes for each city type in the table and analyze for each
rural = combined_data.loc[combined_data['type']=='Rural']
suburban = combined_data.loc[combined_data['type']=='Suburban']
urban = combined_data.loc[combined_data['type']=='Urban']

rural_count = rural.groupby('city')['ride_id'].count()
rural_avg = rural.groupby('city')['fare'].mean()

suburban_count = suburban.groupby('city')['ride_id'].count()
suburban_avg = suburban.groupby('city')['fare'].mean()

urban_count = urban.groupby('city')['ride_id'].count()
urban_avg = urban.groupby('city')['fare'].mean()

urban_avg.head()
```

```python
#Plot data to analyze ride count and average type for each city type in the data
plt.scatter(rural_count,rural_avg,alpha=0.5,edgecolors='black',cmap='gold',s=combined_data['driver_count']*3)
plt.scatter(suburban_count,suburban_avg,alpha=0.5,edgecolors='black',cmap='light sky blue',s=combined_data['driver_count']*3)
plt.scatter(urban_count,urban_avg,alpha=0.5,edgecolors='black',cmap='light coral',s=combined_data['driver_count']*3)

plt.title('Pyper Ride Sharing Data 2016')
plt.xlabel('Total Rides per City')
plt.ylabel('Average Fare ($)')
plt.grid(True)
plt.show()
```

```python
#Group data by city type to analyze total for each city type
city_type = combined_data.groupby('type')
total_fares = city_type['fare'].sum()
total_rides = city_type['ride_id'].count()
total_drivers = city_type['driver_count'].sum()
total_drivers
```

```python
#Pie Chart showing total fares for each city type
labels = total_fares.index
colors=['gold','lightskyblue','lightcoral']
plt.pie(total_fares,labels=labels,autopct='%1.1f%%',shadow='True',colors=colors)
plt.axis('equal')
plt.title('Total Fares by City Type')
plt.show()
```

```python
#Pie Chart showing total rides for each city type
labels = total_rides.index
colors=['gold','lightskyblue','lightcoral']
plt.pie(total_rides,labels=labels,autopct='%1.1f%%',shadow='True',colors=colors)
plt.axis('equal')
plt.title('Total Rides by City Type')
plt.show()
```

```python
#Pie Chart showing total drivers for each city type
labels = total_drivers.index
colors=['gold','lightskyblue','lightcoral']
explode = (0.5,0,0)
plt.pie(total_drivers,labels=labels,autopct='%1.1f%%',shadow='True',colors=colors,explode=explode)
plt.axis('equal')
plt.title('Total Drivers by City Type')
plt.show()
```
